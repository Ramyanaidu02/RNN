```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('spam.csv',encoding='latin-1')
```

```python
data.head()
```

```python
data.rename(columns={'v1':'target','v2':'text'},inplace=True)
```

```python
data.head()
```

```python
vectorizer = CountVectorizer(stop_words='english')
```

```python
X = vectorizer.fit_transform(data['text'])
```

```python
X
```

```python
feature_matrix = X.toarray()
```

```python
feature_matrix
```

```python
print(vectorizer.get_feature_names_out())
```

```python
df_features = pd.DataFrame(feature_matrix,columns=vectorizer.get_feature_names_out())
print(df_features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test= train_test_split(feature_matrix,data['target'],test_size= 0.2,random_state=42)
```

```python
sv = SVC(kernel='linear')
```

```python
sv.fit(X_train,y_train)
```

```python
y_pred = sv.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
predict_message = sv.predict(new_message_features)
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,SimpleRNN
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
spam_data = pd.read_csv('spam.csv',encoding='latin-1')
```

```python
spam_data.head()
```

```python
spam_data.rename(columns={'v1':'target','v2':'text'},inplace=True)
```

```python
spam_data= spam_data[['text','target']]
```

```python
le = LabelEncoder()
spam_data['target'] = le.fit_transform(spam_data['target'])
```

```python
spam_data
```

```python
vacob_size = 1000
```

```python
encoded_docs =[one_hot(d,vacob_size)for d in spam_data['text']]
```

```python
max_length=50
padded_docs = pad_sequences(encoded_docs,maxlen=max_length,padding='post')
```

```python
labels = spam_data['target'].values
```

```python
labels
```

```python
X_train,X_test,y_train,y_test = train_test_split(padded_docs,labels,test_size=0.2,random_state=42)
```

```python
y_train.shape
```

```python
#Build RNN
model = Sequential()
model.add(Embedding(vacob_size,100,input_length=max_length))
model.add(SimpleRNN(128))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,batch_size=32)
```

```python
loss,accuracy=model.evaluate(X_test,y_test)
print(f'test accuracy:{accuracy}')
```

```python
def predict_data(model,sentence,vocab_size,max_length):
    encoded_sentence = one_hot(sentence,vocab_size)
    padded_sentence = pad_sequences([encoded_sentence],maxlen=max_length,padding='post')
    prediction = model.predict(padded_sentence)
    return 'ham' if prediction>0.5 else 'spam'
```

```python
new_sentence = "congratulations you won a ticket"
```

```python
print(predict_data(model,new_sentence,vacob_size,max_length))
```
